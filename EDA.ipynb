```python
agri_cleaned=pd.read_csv(r"G:\Guvi-Datascience\Project\AgriData Explorer_Project 2\agridata.csv")
```

```python
#Top 7 RICE PRODUCTION State Data(Bar_plot)
RP=(agri_cleaned.groupby("state_name")["rice_production_1000_tons"].sum().sort_values(ascending=False).head(7).reset_index())
px.bar(
    RP,
    x="state_name",
    y="rice_production_1000_tons",
    title="Rice Production by State",
    labels={"state_name": "State Name", "rice_production_1000_tons": "Rice Production (1000 tons)"}
)

    
```

```python
#Top 5 Wheat Producing States Data(Bar_chart)and its percentage(%)(Pie_chart)
WP = (agri_cleaned.groupby("state_name")["wheat_production_1000_tons"]
      .sum().sort_values(ascending=False).head(5).reset_index())

# Bar chart
plt.figure(figsize=(8,5))
sns.barplot(x="state_name", y="wheat_production_1000_tons", data=WP)
plt.title("Top 5 Wheat Producing States")
plt.xlabel("State")
plt.ylabel("Wheat Production (1000 tons)")


# Pie chart
plt.figure(figsize=(6,6))
plt.pie(WP["wheat_production_1000_tons"], labels=WP["state_name"], autopct="%1.1f%%", startangle=90)
plt.title("Wheat Production Share (%) - Top 5 States")
```

```python
#Oil seed production by top 5 states
OIL = (agri_cleaned.groupby("state_name")["oilseeds_production_1000_tons"]
       .sum().sort_values(ascending=False).head(5).reset_index())

plt.figure(figsize=(8,5))
sns.barplot(data=OIL, x="state_name", y="oilseeds_production_1000_tons")
plt.title("Oilseeds Production by Top 5 States")
plt.xlabel("State")
plt.ylabel("Oilseeds Production (1000 tons)")


```

```python
#Top 7 SUNFLOWER PRODUCTION  State
SUN = (agri_cleaned.groupby("state_name")["sunflower_production_1000_tons"]
       .sum().sort_values(ascending=False).head(7).reset_index())

px.bar(SUN, x="state_name", y="sunflower_production_1000_tons",
             title="Top 7 Sunflower Producing States", color="state_name")

```

```python
#India's SUGARCANE PRODUCTION From Last 50 Years(Line_plot)
SC = (agri_cleaned.groupby("year")["sugarcane_production_1000_tons"]
      .sum().reset_index())

px.line(SC, x="year", y="sugarcane_production_1000_tons",
              title="SUGARCANE PRODUCTION From Last 50 Years)",
              labels={"value": "Production (1000 tons)", "variable": "Crop"})


```

```python
#Rice Production Vs Wheat Production (Last 50y)
RW = (agri_cleaned.groupby("year")[["rice_production_1000_tons", "wheat_production_1000_tons"]]
      .sum().tail(50).reset_index())

px.line(RW, x="year", y=["rice_production_1000_tons", "wheat_production_1000_tons"],
              title="Rice vs Wheat Production (Last 50 Years)",
              labels={"value": "Production (1000 tons)", "variable": "Crop"})


```

```python
#Rice Production By West Bengal Districts
WB_RICE = (agri_cleaned[agri_cleaned["state_name"]=="West Bengal"]
           .groupby("dist_name")["rice_production_1000_tons"]
           .sum().sort_values(ascending=False).reset_index())

px.bar(
    WB_RICE,
    x="dist_name",
    y="rice_production_1000_tons",
    labels={
        "dist_name": "District",
        "rice_production_1000_tons": "Rice Production (1000 tons)"
    },
    title="Rice Production By West Bengal Districts"
)
```

```python
#Top 10 Wheat Production Years From UP
UP_WHEAT = (agri_cleaned[agri_cleaned["state_name"]=="Uttar Pradesh"]
            .groupby("year")["wheat_production_1000_tons"]
            .sum().sort_values(ascending=False).head(10).reset_index())

px.bar(UP_WHEAT, x="year", y="wheat_production_1000_tons",
             title="Top 10 Wheat Production Years - Uttar Pradesh",
             color="wheat_production_1000_tons")

```

```python
#Millet Production (Last 50y) 
millet_data = agri_cleaned[(agri_cleaned['year'] >= 1966) & (agri_cleaned['year'] <= 2017)]

millet_production = millet_data.groupby('year')[['pearl_millet_production_1000_tons', 'finger_millet_production_1000_tons']].sum().reset_index()

millet_chart = px.line(
    millet_production,
    x='year',
    y=['pearl_millet_production_1000_tons','finger_millet_production_1000_tons'],
    title='Millet Production in India (Last 50y)',
    labels={'year': 'Year'}
)

millet_chart.show()
```

```python
#Sorghum Production (Kharif and Rabi) by Region
SORGHUM = (
    agri_cleaned
    .groupby("state_name")[["kharif_sorghum_production_1000_tons", "rabi_sorghum_production_1000_tons"]]
    .sum()
    .reset_index()
)

plt.figure(figsize=(10,6))
sns.barplot(data=SORGHUM, x="state_name", y="kharif_sorghum_production_1000_tons", color="skyblue", label="Kharif")
sns.barplot(data=SORGHUM, x="state_name", y="rabi_sorghum_production_1000_tons", color="orange", label="Rabi", alpha=0.7)

plt.title("Sorghum Production (Kharif and Rabi) by Region")
plt.xlabel("State Name")
plt.ylabel("Sorghum Production (1000 tons)")
plt.legend(title="Season")
plt.xticks(rotation=45, ha="right")
plt.tight_layout()
plt.show()
```

```python
#Top 7 States for Groundnut Production
GROUNDNUT = (agri_cleaned.groupby("state_name")["groundnut_production_1000_tons"]
             .sum().sort_values(ascending=False).head(7).reset_index())

px.bar(GROUNDNUT, x="state_name", y="groundnut_production_1000_tons",
             title="Top 7 States for Groundnut Production", color="state_name")

```

```python
#Soybean Production by Top 5 States and Yield Efficiency 
import plotly.graph_objects as go

soybean_data = agri_cleaned.groupby('state_name')[['soyabean_production_1000_tons', 'soyabean_yield_kg_per_ha']].sum().reset_index()
soybean_top5 = soybean_data.sort_values(by='soyabean_production_1000_tons', ascending=False).head(5)

soybean_chart = go.Figure()

soybean_chart.add_trace(go.Bar(
    x=soybean_top5['state_name'],
    y=soybean_top5['soyabean_production_1000_tons'],
    name='Production (1000 Tons)',
    text=soybean_top5['soyabean_production_1000_tons'],
    textposition='outside'
))

soybean_chart.add_trace(go.Scatter(
    x=soybean_top5['state_name'],
    y=soybean_top5['soyabean_yield_kg_per_ha'],
    mode='lines+markers',
    name='Yield Efficiency (kg/ha)',
    yaxis='y2'
))

soybean_chart.update_layout(
    title='Top 5 Soybean Producing States & Yield Efficiency',
    xaxis_title='State',
    yaxis_title='Soybean Production (1000 Tons)',
    yaxis2=dict(
        title='Yield Efficiency (kg/ha)',
        overlaying='y',
        side='right'
    )
)

soybean_chart.show()
```

```python
#Oilseed Production in Major States
OIL2 = (agri_cleaned.groupby("state_name")["oilseeds_production_1000_tons"]
        .sum().sort_values(ascending=False).head(10).reset_index())

px.treemap(OIL2, path=["state_name"], values="oilseeds_production_1000_tons",
                 title="Oilseed Production in Major States")
```

```python
#Impact of Area Cultivated on Production (Rice, Wheat, Maize) - doubt
subset = agri_cleaned[["rice_area_1000_ha", "rice_production_1000_tons",
                       "wheat_area_1000_ha", "wheat_production_1000_tons",
                       "maize_area_1000_ha", "maize_production_1000_tons"]]

sns.pairplot(subset)
plt.suptitle("Impact of Area Cultivated on Production (Rice, Wheat, Maize)", y=1.02)
```

```python
#Rice vs. Wheat Yield Across States 
yield_long = pd.melt(
    agri_cleaned[['state_name', 'rice_yield_kg_per_ha', 'wheat_yield_kg_per_ha']],
    id_vars='state_name',
    value_vars=['rice_yield_kg_per_ha', 'wheat_yield_kg_per_ha'],
    var_name='crop',
    value_name='yield'
)

# Rename crop labels for clean legend
yield_long['crop'] = yield_long['crop'].replace({
    'rice_yield_kg_per_ha': 'Rice',
    'wheat_yield_kg_per_ha': 'Wheat'
})

# Plot
rice_wheat_yield_plot = px.scatter(
    yield_long,
    x='state_name',
    y='yield',
    color='crop',   
    title='Rice vs Wheat Yield Across States',
    labels={'yield': 'Yield (kg/ha)', 'state_name': 'State'}
)

rice_wheat_yield_plot.show()
```
